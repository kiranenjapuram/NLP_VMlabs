```python
import pandas as pd
import matplotlib.pyplot as plt

#reading the data
data = pd.read_csv('kc_house_data.csv')

#printing the head of the dataset
print(data.head())

#creating a bar plot
plt.bar(data['bedrooms'],data['price'],label='bedrooms vs price')
plt.xlabel('bedrooms')
plt.ylabel('price')
plt.title('bedrooms vs price')
plt.legend()
plt.show()

#creating a line plot
plt.plot(data['bedrooms'],data['price'],label='bedrooms vs price')
plt.xlabel('bedrooms')
plt.ylabel('price')
plt.title('bedrooms vs price')
plt.legend()
plt.show()

#creating a scatter plot
plt.scatter(data['bedrooms'],data['price'],label='bedrooms vs price')
plt.xlabel('bedrooms')
plt.ylabel('price')
plt.title('bedrooms vs price')
plt.legend()
plt.show()
```
